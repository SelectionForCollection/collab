```python
import pandas as pd
data = pd.read_csv("titanic.csv")
data.info()
```

```python
data.head()
```

```python
import plotly.express as px
import plotly.graph_objs as go
graph = go.Figure()
color = ['red' if _ == 'male' else 'blue' for _ in data.Sex]
marker = {'line': {'color': 'black', 'width': 2}}
graph.add_trace(go.Bar(x=data.Age, y=data.Survived, marker_color=color, marker=marker))
graph.update_xaxes(title='Возраст', title_font=dict(size=16), tickangle=315, tickfont=dict(size=14))
graph.update_yaxes(title='Количество выживших', title_font=dict(size=16), tickfont=dict(size=14))
graph.update_layout(
    title={
        "text": "Выжившие на титанике",
        "x": 0.5,
        "font": {
            "size": 20,
        }
    },
    height=700,
    #margin=dict(l=0, r=0, t=0, b=0),
    xaxis_showgrid=True,
    yaxis_showgrid=True,
    xaxis_gridwidth=2,
    yaxis_gridwidth=2,
    xaxis_gridcolor='ivory',
    yaxis_gridcolor='ivory'
)


graph.show()
```

```python
pie = go.Figure()
graph_pie = go.Pie(
    labels=data.Sex,
    values=data.Survived,
    marker=dict(line=dict(color='black', width=2))
)
pie.update_layout(
    title={
        "text": "Выжившие на титанике",
        "x": 0.5,
        "font": {
            "size": 20,
        }
    },
    height=700,
)
pie.add_trace(graph_pie)
pie.show()
```

```python
import matplotlib.pyplot as plt
buff = [[data.Age[i], data.Survived[i]] for i in range(len(data.Age))]

counter = 0
piff_buff = []
for age in data.Age:
  for el in buff:
    if el[0] == age and el[1] == 1:
      counter += 1
  piff_buff.append([age, counter])
  counter = 0

piff_buff.sort()

age = [i[0] for i in piff_buff]
survived = [i[1] for i in piff_buff]
plt.plot(age, survived, color='crimson', marker='o', markersize=7,
         markerfacecolor='white', markeredgewidth=2, markeredgecolor='black')

plt.title('Зависимость выживаемости от возраста')
plt.xlabel('Возраст')
plt.ylabel('Кол-во выживших')

plt.grid(True, linewidth=2, color='mistyrose')
plt.show()
```
